**Email Spam Detection**

```python
# importing necessary dependencies

import pandas as pd 
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("mail_data.csv")
```

```python
print(df.head(10))
```

```python
data = df.where((pd.notnull(df)),'')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.loc[data['Category'] == 'spam', 'Category',] = 0
data.loc[data['Category'] == 'ham', 'Category',] = 1
```

```python
# separating Categories and Meassages
X = data['Message']
Y = data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words= 'english', lowercase = True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
# Training and Testing messages in textual form

print("################### Training Data ######################")
print(X_train)
print("\n################### Testing Data ######################")
print(X_test)
```

```python
# Training and Testing messages in matrix form using TFIDF

print("################### Training Data ######################")
print(X_train_features)
print("\n################### Testing Data ######################")
print(X_test_features)
```

```python
model = LogisticRegression()
model.fit(X_train_features, Y_train)
```

```python
predicted_output_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, predicted_output_on_test_data)

print(accuracy_on_test_data * 100,'%')
```

```python


your_mail = ['this is 2nd time we have tried to contact u. U have won the AE400 prize. 2 claim is easy, just call 0980932423 NOW! only 10p per  minute. BT-national-rate']

your_mail_features = feature_extraction.transform(your_mail)
predicted_output_on_your_mail = model.predict(your_mail_features)

for i in range(len(your_mail)):
    if predicted_output_on_your_mail[i] == 0:
        print("Spam Mail")
        # print(your_mail_features)
    else:
        print("Ham Mail")
        # print(your_mail_features)
```
